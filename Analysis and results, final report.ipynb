# Data modeling

In this part we tried to look for correlations in our data. Looking at the above plotted graph, we already have a sense that there won't be a strong correlation in our data.

Our target feature is the total number of reports per area. We will try to predict that feature given the wealth of that area.

First, we want to create a data frame that shows the number of reported issues per area.

## Preparing the data

```python
#Take useful data from the FixYourStreet dataFrame
df1 = df_fixyourstreet_data[["Postal Code", "Category_title"]]
```

```python
#Get dummies for the types of  street issues
cat_dum  = pd.get_dummies(df1['Category_title'])
```

```python
#Concatenate both data frames. This dataFrame will be used to get the sum of all issues per areas.
df1 = pd.concat([df1, cat_dum], axis=1)
```

```python
#Get all issues per area

df1 = df1.groupby(by=['Postal Code'], as_index=False).sum()

df1
```

```python
#Merge data frames to get all relevant data in one single data frame.
df_pb_per_area = pd.merge(df_median_housing_price, df1, how='left', on = 'Postal Code')

df_pb_per_area
```

```python
#Trim columns
df_pb_per_area = df_pb_per_area.drop('Unnamed: 0', 1)

#Add total number of issues per areas
#First we need to select the columns we want to sum
column_list = list(df_pb_per_area)
#We remove these 2 columns because we don't want to sum them
column_list.remove('Median housing price')
#Sum the required columns
df_pb_per_area['total report'] = df_pb_per_area[column_list].sum(axis=1)

df_pb_per_area
```

```python
#Updating columns name
df_pb_per_area = df_pb_per_area.rename(columns={'Median housing price': 'Median_housing_price',
                                                'Postal Code':'Postal_code',
                                                'wealth category': 'wealth_category', 
                                                'Leaks and Drainage': 'Leaks_and_Drainage',
                                                'Road or path defects':'Road_or_path_defects',
                                               'Litter and Illegal Dumping':'Litter_and_Illegal_Dumping',
                                               'Street Lighting':'Street_Lighting',
                                               'Tree and Grass Maintenance':'Tree_and_Grass_Maintenance',
                                               'total report':'total_report'})

df_pb_per_area.head()
```

```python
#Get dummies for the wealth category

dummy_wealth = pd.get_dummies(df_pb_per_area['wealth_category'])

df_pb_per_area = pd.concat([df_pb_per_area, dummy_wealth], 1)

df_pb_per_area = df_pb_per_area.rename(columns={'relatively unwealthy': 'relatively_unwealthy',
                                                'relatively wealthy':'relatively_wealthy'})

df_pb_per_area['wealth_category'] = df_pb_per_area['wealth_category'].astype('category')

df_pb_per_area.head()
```

## Linear regression

We have only chosen to undertake the regression's to prove that the information that we have gathered has no predictive use. Our initial goal was more focused on data analysis and seeing if there was a correlation between the relative wealth of an area and the number/type of fault/response time.

We have chosen Total_Report (the total number of incidents) as the target feature. We are interested in whether we can predict the number of reports in a given area. This could be useful for budgeting within a council district if it was accurate. Resources could be better allocated, etc.

### Training the data

First, we will try to predict the total number of reports only taking the feature "median housing price".

```python
#Import statsmodels package for training a linear regression model.
import statsmodels.formula.api as sm

# The formula specifies the target feature and the descriptive features used for training the 
# linear regression model.
# We first train a simple linear regression with only one descriptive feature (Size).
# For training the model we call the method fit() on the given data stored in our df dataframe.
lm = sm.ols(formula="total_report ~  Median_housing_price", data=df_pb_per_area).fit()

print(lm.params)
```

We can also see from the weights that the prediction model doesn't seem really strong (the weight for the median_housing_price is really low). In order to check whether there is any statistical significance to this model, we will print its summary.

```python
print(lm.summary())
```

We can see from the report that the model learnt is not statistically significant. We are therefore unlikely to get any good result from it. For the purpose of completing our data analysis, we will still use it to predict our target feature, and we will check its error metrics.

### Plotting the model

```python
#WFirst we need the maximum and minimum median housing price
X_minmax = pd.DataFrame({'Median_housing_price': [df_pb_per_area.Median_housing_price.min(), df_pb_per_area.Median_housing_price.max()]})
X_minmax.head()
```

```python
# Make predictions for the min and max median housing price
predictions = lm.predict(X_minmax)
predictions
```

```python
#Plot the observed data against the learnt regression line
df_pb_per_area.plot(kind='scatter', x='total_report', y='Median_housing_price')
plt.plot(predictions, X_minmax, c='red', linewidth=2)
```

### use the model to predict our target feature

```python
predictions = lm.predict(df_pb_per_area)

predictions
```

### Evaluate model

```python
#Pair the actual and the predicted values
print("Actual - Predicted:\n", (df_pb_per_area.total_report - lm.predict(df_pb_per_area)))
print("\n(Actual - Predicted) squared:\n", (df_pb_per_area.total_report - lm.predict(df_pb_per_area))**2)
```

```python
# Print the Mean Squared Error of the model on the training set
mse = ((df_pb_per_area.total_report - lm.predict(df_pb_per_area))** 2).mean()
print("\nMean Squared Error:\n", mse)
```

We are now certain that this model isn't good for prediction the results: the error metrics are really high. We will try other prediction models to see if they are better.

### Train a new model

```python
# Training the model
lm = sm.ols(formula="total_report ~  Median_housing_price + C(wealth_category)", data=df_pb_per_area).fit()

# Print the weights
print(lm.params)
```

```python
print(lm.summary())
```

This new model seems slightly better, even though there it is still not very statistically significant. We will try to predict with it and have a look at its error metrics.

```python
df_classif.corr()
```
